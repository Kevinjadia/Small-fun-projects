# Consider the below snake and ladder board to track your position while playing this game

snake_and_ladder.png

```python
import random
pos1=1  # position of player 1
pos2=1  # position of player 2
snakes_ladders={1:38,4:14,8:30,28:76,21:42,50:67,71:92,88:99,97:78,95:56,88:24,62:18,48:26,36:6,32:10}

p1=input("\nEnter name of player 1")
p2=input("\nEnter name of player 2")

while pos1<=100 and pos2<=100:
    
    if pos1==100:
        print("**********************************************************")
        print("\n",p1," won!!!")
        print("**********************************************************")
        
    elif pos2==100:
        print("**********************************************************")
        print("\n",p2," won!!!")
        print("**********************************************************")
    
    else:
        print("*************************************************************************")
        
        print("\n",p1,"'s turn")
        temp1=input("\n Press enter to roll die")
        die=random.randint(1, 6)
        print("\n Value on die:",die)
        
        if die+pos1<=100:
            pos1=pos1+die
        
        if pos1 in snakes_ladders:
            pos1=snakes_ladders[pos1]
        print("\n",p1,"'s Position:",pos1)
        
        print("------------------------------------------------------------------------")
        
        print("\n",p2,"'s turn")
        temp2=input("\n Press enter to roll die")
        die=random.randint(1, 6)
        print("\n Value on die:",die)
        
        if pos2+die<=100:
            pos2=pos2+die
        
        if pos2 in snakes_ladders:
            pos2=snakes_ladders[pos2]
        print("\n",p2,"'s Position:",pos1)

        print("*************************************************************************")
```
